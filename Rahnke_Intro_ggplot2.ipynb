### Fundamental Structure of `ggplot2` Expression

```R
ggplot(data = <DATA>) + 
  <GEOM_FUNCTION>(mapping = aes(<MAPPINGS>))
```

`<>` refers to values you need to define.


### Fundamental Structure of `ggplot2` Expression - cont'd

`ggplot()`:

 * creates a coordinate system that you can add layers to.

 * `ggplot2` comes with many `geom` functions that each add a different type of layer to a plot. E.g.:

    * `geom_point`: scatter plots 
* `aes` stands for aesthetics (e.g.: size, the shape, or the color of your points)
  * `aes` uses vector while `geom_point` param takes single values
* Style can be modified using `params`

### `GEOM_FUNCTION`

 * A **geom** is the geometrical object that a plot uses to represent data.
    * Visual object in this plot 
 * Plots types in ggplots2 different based on **geoms**.
   * e.g.: `geom_point`, `geom_bar`, `geom_line`, `geom_boxplot`, etc.

### `aes`

* To map an aesthetic to a variable, associate the name of the aesthetic to the name of the variable inside `aes()`. 
  * ggplot2 will automatically assign a unique level of the aesthetic (here a unique color) to each unique value of the variable

````r
# Returns a numerical encoding for the car class

convert_class_int <- function(x) {
  if (x == "2seater") {
    1
  }else if (x == "compact"){
    2
  }else if (x == "subcompact"){
    3
  }else if(x == "midsize" ){
    4
  }else if(x == "midsize"){
    5
  }else if(x == "suv"){
    6
  }else if(x == "minivan"){
    7
  }else if(x == "pickup"){
    8
  }else{
    # Used for defensive programming purposes
    stop("value passed not recognized")
  }
}
````

````r
# assertions used for testing (defensive programming purposes)
pickup_val <- convert_class_int("2seater")
assertthat::assert_that(pickup_val == 1)
````

````r
pickup_val <- convert_class_int("pickup") 
assertthat::assert_that(pickup_val == 8)
````

# Using the `sapply` Function

* The `sapply` function is part of the apply() family 
  * A convenient and `s`implified way to apply a function to the elements of a vector
* Gives each element of the vector to the function 
  * combines the returned values into a new list.




````r
someData <- c(1,2,3,4,5)
power_3 <- function(x){
 x^3
}
processed_data = sapply(someData, power_3)
print(processed_data)
````

### Using the `sapply` Function

* Way to implement a data processing paradigm called `Split`-`Apply`- `Combine`
  * Split: here, `sapply` splits the data into single elements
  * Apply: we apply a function to each element of the data
    * The input is passed in each iteration
  * The data resulting from each function call is combined into a new vector of the same size as the origina 

````r
x = c(1, 2, 3, 4, 5)
names(x) = c("A", "B", "A", "A", "B")
print(x)
````

### `geom` params versus `aes`
* All the params of `geom` can also be params of `aes`
* `aes` mapping typically conveys information about the variable
  * In `aes` it's a way to display a third dimension to the variable of sorts
* In `geom`, the mapping is simply stylistic
* You can combine mappings as long as they are not contradictory

### Uisng Facets

 * Subplots of a larger plots
   * Rach one contains a subset of points as described by a formula
     * Forumula is a data structure not equation
   * Points uniquely assigned to one of the subplots
    
 *  `facet_wrap()` 
       *  formula: `~ varname`
          * read as: `as a funciton of varname`
       * More when we talk about modeling

   * `facet_grid()`
     * formula: two`varname_1 ~  varname_2`
